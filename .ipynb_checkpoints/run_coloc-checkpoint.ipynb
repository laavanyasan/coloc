```r
## Running COLOC on PCOS data from Day et al (2018) and eQTL data from HPO axis. Gtex datasets

#clear workspace
rm(list=ls())

#load packages
library(gdata)
library(ggplot2)
library(RColorBrewer)
library(dplyr)
library(tidyr)
library(coloc)
```

```r
#GEtting my two datasets
GWASdata <- read.table("Dayetal2018GWAS.txt", sep="\t", header=TRUE)
ovary.eQTL <- read.table("condensed_Ovary_GTex_eqtl.txt", sep="\t", header=TRUE)\
pitutary.eQTL <- read.table("condensed_Ovary_GTex_eqtl.txt", sep="\t", header=TRUE)
hypothalamus.eQTL <- read.table("condensed_Ovary_GTex_eqtl.txt", sep="\t", header=TRUE)

head(GWASdata)
head(joined.eQTL)
```

```r
#Running coloc : 
#Obtainerd N and s from Hayes et al 2015. 
# N = tot no of samples, s = proportion of samples that are cases. 
#N for Ovary eQTL : https://www.gtexportal.org/home/eqtls/tissue?tissueName=Ovary

my.res <- coloc.abf(dataset1=list(beta=GWASdata$beta, varbeta=GWASdata$varbeta, N=3948,s = 0.24924, pvalues = GWASdata$p.values, MAF = GWASdata$MAF, snp = GWASdata$snp, type="cc"),
                    dataset2=list(pvalues = joined.eQTL$pval_beta, snp = joined.eQTL$rs_id_dbSNP151_GRCh38p7, MAF = joined.eQTL$maf, type = "quant", N = 180))
my.res
```

```r
#setting GWAS MAF as numeric

GWASdata$MAF <- as.numeric(GWASdata$MAF)
```
