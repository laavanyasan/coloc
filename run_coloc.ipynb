```r
## Running COLOC on PCOS data from Day et al (2018) and eQTL data from HPO axis. Gtex datasets

#clear workspace
rm(list=ls())

#load packages
library(gdata)
library(ggplot2)
library(RColorBrewer)
library(dplyr)
library(tidyr)
library(coloc)
```

```r
#GEtting my two datasets
GWASdata <- read.table("Dayetal2018GWAS.txt", sep="\t", header=TRUE)
ovary.eQTL <- read.table("condensed_Ovary_GTex_eqtl.txt", sep="\t", header=TRUE)
pitutary.eQTL <- read.table("condensed_pituitary_eqtl.txt", sep="\t", header=TRUE)
hypothalamus.eQTL <- read.table("condensed_hypothalamus_eqtl.txt", sep="\t", header=TRUE)

head(GWASdata)
head(ovary.eQTL)
head(pitutary.eQTL)
head(hypothalamus.eQTL)
```

```r
#Running coloc for PITTY!
#Obtainerd N and s from Day et al 2018 
# N = tot no of samples, s = proportion of samples that are cases. 
#N = 113,238
# s= 0.0890
#N for Ovary eQTL : https://www.gtexportal.org/home/eqtls/tissue?tissueName=Ovary N= 237

pit.res <- coloc.abf(dataset1=list(beta=GWASdata$beta, varbeta=GWASdata$varbeta, N=113238,s = 0.0890, pvalues = GWASdata$p.values, MAF = GWASdata$maf, snp = GWASdata$snp, type="cc"),
                    dataset2=list(pvalues = pitutary.eQTL$pval_beta, snp = pitutary.eQTL$rs_id_dbSNP151_GRCh38p7, MAF = pitutary.eQTL$maf, type = "quant", N = 237))
pit.res
```

```r
#Running coloc for OVARY!
#Obtainerd N and s from Day et al 2018 
# N = tot no of samples, s = proportion of samples that are cases. 
#N = 113,238
# s= 0.0890
#N for Ovary eQTL : https://www.gtexportal.org/home/eqtls/tissue?tissueName=Ovary N= 167

ovary.res <- coloc.abf(dataset1=list(beta=GWASdata$beta, varbeta=GWASdata$varbeta, N=113238,s = 0.0890, pvalues = GWASdata$p.values, MAF = GWASdata$maf, snp = GWASdata$snp, type="cc"),
                    dataset2=list(pvalues = ovary.eQTL$pval_beta, snp = ovary.eQTL$rs_id_dbSNP151_GRCh38p7, MAF = ovary.eQTL$maf, type = "quant", N = 167))
ovary.res
```

```r
#Running coloc for Hypothalamus!
#Obtainerd N and s from Day et al 2018 
# N = tot no of samples, s = proportion of samples that are cases. 
#N = 113,238
# s= 0.0890
#N for Hypo eQTL : https://www.gtexportal.org/home/eqtls/tissue?tissueName=Ovary N= 170

hyp.res <- coloc.abf(dataset1=list(beta=GWASdata$beta, varbeta=GWASdata$varbeta, N=113238,s = 0.0890, pvalues = GWASdata$p.values, MAF = GWASdata$maf, snp = GWASdata$snp, type="cc"),
                    dataset2=list(pvalues = hypothalamus.eQTL$pval_beta, snp = hypothalamus.eQTL$rs_id_dbSNP151_GRCh38p7, MAF = hypothalamus.eQTL$maf, type = "quant", N = 170))
hyp.res
```

```r
#Saving res data as text
df.res <- ovary.res$results

df.res

```

```r
lapply(df.res, write, "test.txt", append=TRUE)
```
